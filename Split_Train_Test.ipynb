```python
import os
import shutil
import pandas as pd
from sklearn.model_selection import train_test_split

# Define source and destination directories
source_images_dir = 'C:/Users/bhuhp/Supriya/PACGAN_NEW/Training/data/train_raw'
train_images_dir = 'C:/Users/bhuhp/Supriya/PACGAN_NEW/Training/data/train_source'
test_images_dir = 'C:/Users/bhuhp/Supriya/PACGAN_NEW/Training/data/test_source'

def copy_and_rename_images(data, source_dir, dest_dir):
    if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)
    
    # List files in the source directory for debugging
    source_files = os.listdir(source_dir)
    print(f"Files in source directory ({source_dir}): {source_files}")

    # Create a dictionary to map original IDs to new IDs
    id_mapping = {}
    
    for new_id, (_, row) in enumerate(data.iterrows(), start=0):
        image_id = str(row['ID']) + '.png'  # Ensure correct file extension
        src_path = os.path.join(source_dir, image_id)
        
        # Generate new filename and path (with extension)
        new_filename = f"{new_id}.png"  # Sequential naming with extension
        dst_path = os.path.join(dest_dir, new_filename)
        
        # Print source and destination paths for debugging
        print(f"Source Path: {src_path}")
        print(f"Destination Path: {dst_path}")
        
        if image_id in source_files:
            shutil.copy(src_path, dst_path)
            id_mapping[new_id] = row['Class']  # Use new_id as key
        else:
            print(f"Warning: {src_path} does not exist in source directory and will be skipped.")
    
    return id_mapping

# Load the CSV file containing IDs and class labels
csv_file = 'C:/Users/bhuhp/Supriya/PACGAN_NEW/Training/data/labels.csv'
data = pd.read_csv(csv_file)

# Split the data into training and testing sets
test_size = 0.2  # 20% for testing
train_data, test_data = train_test_split(data, test_size=test_size, stratify=data['Class'], random_state=42)

# Save the split data to new CSV files (optional)
train_data.to_csv('train_data.csv', index=False)
test_data.to_csv('test_data.csv', index=False)

# Copy and rename images for training and testing sets
train_id_mapping = copy_and_rename_images(train_data, source_images_dir, train_images_dir)
test_id_mapping = copy_and_rename_images(test_data, source_images_dir, test_images_dir)

# Generate label CSV files with IDs as integers (no file extension)
train_labels = pd.DataFrame(list(train_id_mapping.items()), columns=['ID', 'Class'])
test_labels = pd.DataFrame(list(test_id_mapping.items()), columns=['ID', 'Class'])

# Save label CSV files
train_labels.to_csv('C:/Users/bhuhp/Supriya/PACGAN_NEW/Training/data/train_labels.csv', index=False)
test_labels.to_csv('C:/Users/bhuhp/Supriya/PACGAN_NEW/Training/data/test_labels.csv', index=False)
```

```python
import os
import shutil
from PIL import Image

def create_folders_for_sizes(base_dir, sizes):
    """
    Creates folders for each specified size in the base directory.
    """
    for size in sizes:
        folder_name = f'Real{size}x{size}'
        folder_path = os.path.join(base_dir, folder_name)
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)
            print(f"Created folder: {folder_path}")

def resize_and_move_images(source_dir, sizes):
    """
    Resizes images and moves them to folders based on their size.
    """
    # List all files in the source directory
    source_files = os.listdir(source_dir)
    
    # Create folders for each size
    create_folders_for_sizes( 'C:/Users/bhuhp/Supriya/PACGAN_NEW/Training/data/test_final', sizes)
    
    for file_name in source_files:
        src_path = os.path.join(source_dir, file_name)
        
        try:
            with Image.open(src_path) as img:
                for size in sizes:
                    # Resize image to the target size using Lanczos resampling
                    resized_img = img.resize((size, size), Image.LANCZOS)
                    
                    # Determine target directory and path
                    size_folder = f"Real{size}x{size}"
                    target_dir = os.path.join('C:/Users/bhuhp/Supriya/PACGAN_NEW/Training/data/test_final', size_folder)
                    dst_path = os.path.join(target_dir, file_name)
                    
                    # Save the resized image
                    resized_img.save(dst_path)
                    print(f"Resized and saved {file_name} to {dst_path}")
                    
        except Exception as e:
            print(f"Error processing {src_path}: {e}")

# Example usage
source_images_dir = 'C:/Users/bhuhp/Supriya/PACGAN_NEW/Training/data/test_source'
sizes = [4, 8, 16, 32, 64, 128, 256]

# Resize images and move to size-specific folders
resize_and_move_images(source_images_dir, sizes)
```
